```python
import pandas as pd
import numpy as np
```

```python
view_count_df = pd.read_csv('videoList_final_2.csv').loc[:,['영화 제목', '조회수']]
view_count_df.rename(columns={'영화 제목': '영화명'}, inplace=True)
view_count_df
```

```python
view_count_sum = view_count_df.groupby('영화명').apply(lambda x: np.sum(x['조회수']))
view_count_mean = view_count_df.groupby('영화명').apply(lambda x: np.mean(x['조회수']))
```

```python
view_count_stat = pd.DataFrame(columns=['영화명', '조회수 합', '조회수 평균'])
view_count_stat['영화명'] = view_count_sum.index
view_count_stat['조회수 합'] = view_count_sum.values
view_count_stat['조회수 평균'] = view_count_mean.values
view_count_stat
```

```python
view_count_stat.to_csv('view_count_state_2.csv', encoding='utf-8-sig')
```
